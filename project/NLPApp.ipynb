```python
import nlpcloud

class NLPApp : 

    def __init__(self): 
        self.__database = {}
        self.menu()

    def menu(self) : 
        user_input = input("""
        HELLO! & Wellcom in NLPApp 
        
        1.Resgister
        2.Already a member? Login! 
        3.exit(aaye hi kyun the phir?)
        
        """)
        if user_input == "1": 
            self.register()
        elif user_input == "2": 
            self.login()
        else: 
            exit()
            
    def register(self ): 
        print("Registration:")
        name = input("enter name :")
        email = input("enter email: ")
        password = input("enter password")

        if email in self.__database: 
            print("user already exist . login")
            self.login()
        else: 
            self.__database[email] = [name, password]
            print ( "registered succesfully")
            #print(self.__database)
        self.login()
        
    def login(self): 
        print("\nnow login")
        email = input("enter email: ")
        password = input("enter password")

        if email not in self.__database:
            print("user does not exist, register first\n")
            return self.register()
        else:
            if self.__database[email][1] == password:
                print("login successfully")
                return self.second_menu()
            else:
                print("wrong password\ntryagain\n")
                return self.login()
                
    def second_menu(self): 
        user_input = input("""
        Hello! Please select the function you would like to proceed with:
        1. NER (entity extraction).
        2. Grammer and spelling correction.
        3. summarization.
        4. Headline genrater. 
        5. Return to home page.
        6. Exit to the program.
        """)

        if user_input == "1": 
            self.NER()
        elif user_input ==  "2": 
            self.grammer_correction()
        elif user_input == "3": 
            self.summarization()
        elif user_input == "4": 
            self.headline()
        elif user_input == "5":
            self.menu()
        else: 
            return 0

    def NER(self): 
        para = input("enter the paragraph you want to check: ")
        search = input("enter the searching word")
        client = nlpcloud.Client("gpt-oss-120b", "1a260e5d955b2933c02a9d654de116e8a78ab16a", gpu=True)
        response = client.entities(para,searched_entity=search)
        l =[]
        for i in range(len(response["entities"])): 
            text = (response["entities"][i]["text"]) 
            l.append(text)
        print(l)
        self.second_menu()
        
    def grammer_correction(self): 
        para = input("enter the paragraph you want to correct: ")
        client = nlpcloud.Client("gpt-oss-120b", "1a260e5d955b2933c02a9d654de116e8a78ab16a", gpu=True)
        responce = client.gs_correction(para)
        print("correct version is : ", responce)
        self.second_menu()
        
    def summarization(self): 
        para = input("enter the paragraph you want to summarize: ")
        size = input("enter the formate small/medium/large: ")
        client = nlpcloud.Client("gpt-oss-120b", "1a260e5d955b2933c02a9d654de116e8a78ab16a", gpu=True)
        response = client.summarization(para,size=size)
        print(response)
        self.second_menu()
        
    def headline(self):
        para = input("enter the paragraph to genrate the headline: ")
        client = nlpcloud.Client("t5-base-en-generate-headline", "1a260e5d955b2933c02a9d654de116e8a78ab16a", gpu=False)
        response = client.summarization(para) 
        print("HEADLINE\n",response)
        self.second_menu()

        
app = NLPApp()
```
